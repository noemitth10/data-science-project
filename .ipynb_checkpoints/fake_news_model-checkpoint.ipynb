```python
import pandas as pd
import numpy as np
import itertools
from sklearn.model_selection import train_test_split
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import string as st
import re
import nltk
from nltk import PorterStemmer, WordNetLemmatizer
import matplotlib.pyplot as plt
```

```python
#Load the data from the CSV files
fake_news = pd.read_csv('Fake.csv')
fake_news.isnull().any()
```

```python
true_news = pd.read_csv('True.csv')
true_news.isnull().any()
```

```python
#Load the data into dataframes
df_fake = pd.DataFrame(fake_news)
df_fake
```

```python
df_true = pd.DataFrame(true_news)
df_true
```

```python
labels = []

for i in range(0, len(df_fake)):
    labels.append('FAKE')
```

```python
df_fake['label'] = labels
df_fake
```

```python
labels = []

for i in range(0, len(df_true)):
    labels.append('TRUE')
```

```python
df_true['label'] = labels
df_true
```

```python
#Merge the two dataset into one
frames = [df_fake, df_true]
result = pd.concat(frames)
result
```

```python
#Shuffle the rows of the new dataframe
final_df = result.sample(frac=1).reset_index(drop=True)
final_df
```

```python
final_df.info()
```

```python
final_df.shape
```

```python
print(np.unique(final_df['label']))
print(np.unique(final_df['label'].value_counts()))
```

```python
#Remove punctuations
def remove_punct(text):
    return ("".join([ch for ch in text if ch not in st.punctuation]))

final_df['removed_punc'] = final_df['text'].apply(lambda x: remove_punct(x))
```

```python
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
#Tokenization and making everything lowercase
def tokenize(text):
    text = word_tokenize(text)
    return [x.lower() for x in text]

final_df['tokens'] = final_df['removed_punc'].apply(lambda msg : tokenize(msg))
```

```python
final_df.head()
```

```python
#Removing stop words
nltk.download('stopwords')
```

```python
def remove_small_words(text):
    return [x for x in text if len(x) > 3 ]
```

```python
final_df['filtered_tokens'] = final_df['tokens'].apply(lambda x : remove_small_words(x))
final_df.head()
```

```python
#x = final_df.drop('label', axis=1)
#y = final_df['label']

#x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.9)
```

```python
stopwords = set(nltk.corpus.stopwords.words('english'))
```

```python
len(stopwords)
```

```python
def clean_text(text):
    text = [word for word in text if word not in stopwords]
    return text

final_df['clean_tokens'] = final_df['filtered_tokens'].apply(lambda x : clean_text(x))
```

```python
final_df.head()
```

```python
#Lemmatization
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
word_net = WordNetLemmatizer()

def lemmatize(text):
    return [word_net.lemmatize(word) for word in text]
```

```python
final_df['lemma_words'] = final_df['clean_tokens'].apply(lambda x : lemmatize(x))
```

```python
final_df.head()
```

```python
# Create sentences to get clean text as input for vectors

def return_sentences(tokens):
    return " ".join([word for word in tokens])
```

```python
final_df['clean_text'] = final_df['lemma_words'].apply(lambda x : return_sentences(x))
final_df.head()
```

```python
#Generating word cloud
from wordcloud import WordCloud, ImageColorGenerator

text = " ".join([x for x in final_df['clean_text']])
# Create and generate a word cloud image:
wordcloud = WordCloud(max_font_size=30, max_words=1000).generate(text)

# Display the generated image:
plt.figure(figsize= [20,10])
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
